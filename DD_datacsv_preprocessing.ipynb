```python
import os
import requests
import zipfile
from googleapiclient.discovery import build
from googleapiclient.errors import HttpError
from google.colab import drive

# Mount Google Drive
drive.mount('/content/drive')

# Base URL and target directory on Google Drive
base_url = "https://dcapswoz.ict.usc.edu/wwwdaicwoz/?fbclid=IwZXh0bgNhZW0CMTAAAR37Bv7CdFZQkwq0VmwHy3R7IWqUh3W6hv5-at-1coOp-ixGSqJpMKllsh4_aem_yKRvaX1nT8B8CPLlzf5DFw"
file_names = [f"{i}_P.zip" for i in range(300, 380)]  # Adjust range as per your file list
google_drive_dir = "/content/drive/MyDrive/DAIC_WOZ"

# Ensure the target directory exists
os.makedirs(google_drive_dir, exist_ok=True)

# Function to get available space in Google Drive
def get_available_space():
    try:
        service = build('drive', 'v3')
        about = service.about().get(fields="storageQuota").execute()
        total_space = int(about['storageQuota']['limit'])
        used_space = int(about['storageQuota']['usage'])
        free_space = total_space - used_space
        return free_space
    except HttpError as error:
        print(f"Error checking space: {error}")
        return None

# Function to download and extract
def download_and_extract(file_name, target_dir):
    url = base_url + file_name
    local_zip_path = f"./{file_name}"

    # Check available space before download
    available_space = get_available_space()
    if available_space is not None:
        print(f"Available Google Drive space: {available_space / (1024**2):.2f} MB")
        if available_space < 500 * 1024 * 1024:  # Require at least 500MB free for download and extraction
            print("Insufficient space in Google Drive. Stopping downloads.")
            return False

    # Download file
    print(f"Downloading {file_name}...")
    response = requests.get(url, stream=True)
    with open(local_zip_path, 'wb') as file:
        for chunk in response.iter_content(chunk_size=8192):
            file.write(chunk)

    # Extract zip file
    print(f"Extracting {file_name}...")
    with zipfile.ZipFile(local_zip_path, 'r') as zip_ref:
        zip_ref.extractall(target_dir)

    # Remove zip file after extraction
    os.remove(local_zip_path)
    print(f"{file_name} processed successfully!")
    return True

# Iterate through the files
for file_name in file_names:
    if not download_and_extract(file_name, google_drive_dir):
        break
```

```python
train_csv_path = "/content/drive/MyDrive/DD_final_project/csv/train_set.csv"
validate_csv_path = "/content/drive/MyDrive/DD_final_project/csv/validation_set.csv"
test_csv_path = "/content/drive/MyDrive/DD_final_project/csv/full_test_split.csv"
base_dir = "/content/drive/MyDrive/DAIC_WOZ"
```

```python
train_sessions = pd.read_csv(train_csv_path)['Participant_ID'].astype(str).tolist()
validate_sessions = pd.read_csv(validate_csv_path)['Participant_ID'].astype(str).tolist()
#test_sessions = pd.read_csv(test_csv_path)['Participant_ID'].astype(str).tolist()
```

```python
test_df=pd.read_csv(test_csv_path)
test_df
```

```python
test_sessions = pd.read_csv(test_csv_path)['Participant_ID'].astype(str).tolist()
```

```python
known_sessions = set(train_sessions + validate_sessions + test_sessions)
```

```python
train_dir = os.path.join(base_dir, "train")
validate_dir = os.path.join(base_dir, "validate")
test_dir = os.path.join(base_dir, "test")
```

```python
os.makedirs(train_dir, exist_ok=True)
os.makedirs(validate_dir, exist_ok=True)
os.makedirs(test_dir, exist_ok=True)
```

```python
existing_train_folders = {d for d in os.listdir(train_dir) if os.path.isdir(os.path.join(train_dir, d))}
existing_validate_folders = {d for d in os.listdir(validate_dir) if os.path.isdir(os.path.join(validate_dir, d))}
existing_test_folders = {d for d in os.listdir(test_dir) if os.path.isdir(os.path.join(test_dir, d))}
```

```python
for folder in os.listdir(base_dir):
    folder_path = os.path.join(base_dir, folder)

    # Skip if it's one of the main destination directories
    if folder_path in [train_dir, validate_dir, test_dir]:
        continue

    # We're looking for directories named like "XXX_P"
    if os.path.isdir(folder_path) and folder.endswith("_P"):
        # Extract the session_id by removing the trailing "_P"
        session_id = folder[:-2]  # '123_P' -> '123'

        # Check if folder is already organized
        if folder in existing_train_folders or folder in existing_validate_folders or folder in existing_test_folders:
            # Already organized, skip it
            continue

        # Move according to the session category
        if session_id in train_sessions:
            dest_path = os.path.join(train_dir, folder)
            if not os.path.exists(dest_path):
                shutil.move(folder_path, dest_path)

        elif session_id in validate_sessions:
            dest_path = os.path.join(validate_dir, folder)
            if not os.path.exists(dest_path):
                shutil.move(folder_path, dest_path)

        elif session_id in test_sessions:
            dest_path = os.path.join(test_dir, folder)
            if not os.path.exists(dest_path):
                shutil.move(folder_path, dest_path)

        elif session_id not in known_sessions:
            # Remove unmatched folders
            shutil.rmtree(folder_path)
```
